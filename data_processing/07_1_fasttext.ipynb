```python
from gensim.models.fasttext import FastText as FT_gensim

# Set file names for train and test data
corpus_file = open('corpus.txt', encoding='utf-8')

model = FT_gensim(size=200, iter=4000, window=5)

# build the vocabulary
model.build_vocab(corpus_file=corpus_file)

# train the model
model.train(
    corpus_file='corpus.txt', epochs=model.epochs,
    total_examples=model.corpus_count, total_words=model.corpus_total_words
)
```

```python
#from gensim.test.utils import common_texts

model.wv.save_word2vec_format('ft_0825_7.txt')
model.save('ft_0825_7.model')
result = FT_gensim.load('ft_0825_7.model')
```

```python
d = model.wv.most_similar(positive=[""])
for (x, y) in d:
    print("%s - %f" % (x,y))
```
