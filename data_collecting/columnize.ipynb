```python
import os
import re
import pandas as pd 

xname = os.path.join(os.path.dirname(os.getcwd()), 'xlsx_data', f"{'xlsx_#2_text'}.xlsx")
df = pd.read_excel(xname, index_col=0)
df = df[df['image']==0].reset_index(drop=True)
result_list = []
print(len(df))
for i in range(len(df)):
    print(i)
    print(df.loc[i, 'seqnum'])
    
    seqnum = df.loc[i, 'seqnum']
    cbub_code = df.loc[i, 'cbub_code']
    title = df.loc[i, "title"]
    url = df.loc[i, 'url']
    path = df.loc[i, 'path']
    text = df.loc[i, 'text']
    image = df.loc[i, 'image']
    
    if '' in text:
        appeal = 1
    else:
        appeal = 0
    
    text = re.sub(r"(-\s?\d\s?-)", ' ', text)
    text = re.sub(r' +', ' ', text)
    text = re.sub(r'_{2,}', '', text)
    
    try:
        info = re.search(r'\b \b(.+)', text).group(1) #       .
    except:
        continue
        
    cid = info.split()[0]
    case = re.sub(r"(\d+\w+)", "", info).strip()
    undertrial = re.findall(r'\b  \s(.+)[\n{2,}]+(.+)+', text)
    try:
        prosecutor = re.search(r'\b \s(.+)', text).group(1)
    except:
        prosecutor = 0
    
    try:
        sentence = re.sub(r'\n{2,}', ' ', text[text.index(' ')+3:text.rindex(' ')])
        if '' in text:
            honor = re.findall(r'\b|\b(.+)', text)
            body = text[text.index(' '):text.rindex('')]
        else:
            honor = re.search(r'\b \b(.+)', text).group(1)
            body = text[text.index(' '):text.rindex('')]
    except:
        try:
            sentence = re.search('   .+\n{2,}(.+)', text).group(1)
            if '' in text:
                honor = re.findall(r'\b|\b(.+)', text)
                body = text[text.index(sentence)+len(sentence):text.rindex('')]
            else:
                try:
                    honor = re.search(r'\b \b(.+)', text).group(1)
                    body = text[text.index(sentence)+len(sentence):text.rindex('')]
                except:
                    honor = 0 #     .
                    body = text[text.index(sentence)+len(sentence):]
        except:
            continue #      
    body = body.replace(' ', '').replace('   ', '').replace(' ', '').replace(' ', '').replace(' ', '')
    body = re.sub(r'[\x00-\x08\x0b\x0c\x0e-\x1f\x7f-\xff]', '', body)
    body = re.sub("\n{2,}", "", body)
    body = re.sub(r'(\d+\.\s\d+\.\s\d+\.)(\s\d+:\d+)?', '', body)
    body = re.sub(r'[\(\[].*?[\)\]]', '', body)
    
    new_dict={
        'seqnum': seqnum,
        'cbub_code': cbub_code,
        'title': title,
        'url': url,
        'path': path,
        'text': text, 
        'image': image,
        'appeal': appeal,
        'cid': cid,
        'case': case,
        'undertrial': undertrial,
        'prosecutor': prosecutor,
        'sentence': sentence,
        'honor': honor,
        'body': body
    }
    result_list.append(new_dict)
    
df = pd.DataFrame(result_list)        
save_name = os.path.join(os.path.dirname(os.getcwd()), 'xlsx_data', f"{'xlsx_#3_columnize'}.xlsx")
excel_writer = pd.ExcelWriter(save_name, engine='xlsxwriter')
df.to_excel(excel_writer)
excel_writer.save()
print(len(df))
```
